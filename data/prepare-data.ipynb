```python
import pandas as pd
import numpy as np
import math
import os
import random 
```

```python
df = pd.read_csv("eec/data.csv")
df
```

```python
df = df.dropna(subset=["Emotion"])

label = {"joy" : 1, "sadness" : 0, "fear" : 0, "anger" : 0 }

def change_label(text) :
    return str(label[text])
```

```python
df["polarity"] = df["Emotion"].apply(change_label)

df = df.drop(columns=["ID", "Template", "Person", "Race", "Emotion", "Emotion word"])
```

```python
from sklearn.model_selection import train_test_split
X = df["Sentence"].values
y = df["polarity"].values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123, stratify=y, test_size=0.05)

train = pd.DataFrame(data={0: y_train , 1 : X_train})
test = pd.DataFrame(data={0: y_test , 1 : X_test})
```

```python
train.to_csv("eec/train.csv", index=None, header=None, sep="\t")
test.to_csv("eec/test.csv", index=None, header=None,sep="\t")
```
