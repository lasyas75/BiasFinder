# Bugs User Study

A user study to uncover the false positive from bugs detected


```python
import pandas as pd
import numpy as np
import math
```

#### Read Mutant Original Data

```python
data_dir = "imdb_mutant"
dfm = pd.read_csv("../data/" + data_dir + "/male/test.csv", header=None, sep="\t", names=["label", "text", "template"])
dff = pd.read_csv("../data/" + data_dir + "/female/test.csv", header=None, sep="\t", names=["label", "text", "template"])
```

#### Read Mutant Prediction

```python
def read_txt(fpath):
    pred = []
    file = open(fpath)
    lines = file.readlines()
    for l in lines :
        pred.append(int(l))
    file.close()
    
    return pred
```

```python
output_dir = "trial_on_imdb"

result_dir = "../result/" + output_dir + "/"

rm = result_dir + "results_data_male.txt"
rf = result_dir + "results_data_female.txt"

mpred = read_txt(rm)
fpred = read_txt(rf)

print(len(mpred))
print(len(fpred))
```

#### Append Prediction Result to Data

```python
dfm["pred"] = mpred
dff["pred"] = fpred
```

```python
dfm
```

```python
dfm["template"] = dfm["template"].astype("category")
dfm["template_id"] = dfm["template"].cat.codes

dff["template"] = dff["template"].astype("category")
dff["template_id"] = dff["template"].cat.codes
```

```python
dfm
```

#### Get Data that trigger bugs, means label and prediction are different

```python
mbug = dfm[dfm["label"] != dfm["pred"]]
fbug = dff[dff["label"] != dff["pred"]]
```

```python
mbug
```

```python
fbug
```

#### Group Data by Template Then Sort it

```python
mcount = mbug.groupby("template_id").count()
mcount["#bug"] = mcount["text"]
mcount = mcount.drop(columns=["label", "text", "template", "pred"])
mcount = mcount.sort_values('#bug',ascending=False)
mcount
```

```python
fcount = fbug.groupby("template_id").count()
fcount["#bug"] = fcount["text"]
fcount = fcount.drop(columns=["label", "text", "template", "pred"])
fcount = fcount.sort_values('#bug',ascending=False)
fcount
```
