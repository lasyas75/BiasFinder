```python
# models = ["bert-base-uncased", "bert-base-cased", "roberta-base", "albert-base-v2", "microsoft/mpnet-base", "microsoft/deberta-base", "facebook/muppet-roberta-base", "google/electra-base-generator"]

# predict original
# "roberta-base", "albert-base-v2", "microsoft/mpnet-base", "microsoft/deberta-base", "facebook/muppet-roberta-base", "google/electra-base-generator"

# predict mutants
# "roberta-base", "albert-base-v2", "microsoft/mpnet-base", "microsoft/deberta-base", "facebook/muppet-roberta-base", "google/electra-base-generator"


# models = ["bert-base-uncased"]

models = ["bert-base-uncased", "bert-base-cased"]

# models = ["bert-base-uncased", "bert-base-cased", "roberta-base", "xlnet-base-cased",
#           "albert-base-v2", "microsoft/mpnet-base", "microsoft/deberta-base",
#           "facebook/muppet-roberta-base", "google/electra-base-generator"]

mutation_tool = "biasfinder"
tool_to_bias_type = {"biasfinder": ["gender", "country", "occupation"], "mtnlp": [""]}
task = "imdb"  # dataset used for fine-tuning
mutant = "imdb"  # dataset used for generating mutants


d = pd.DataFrame(columns=["model", "tool", "bias_type", "mutant", "fairness_violation"])

for model in models:
    for mutation_tool, types in tool_to_bias_type.items():
        for bias_type in types :
            # print("Mutation Tool: ", mutation_tool)
            # print("Model: ", model)
            # print("Bias Type: ", bias_type)
            df = load_mutant_and_prediction(mutation_tool, model, bias_type, mutant)
            fairness_violation = evaluate_fairness_violation(
                df["prediction"], df["original_prediction"])

            d = d.append(
                {
                    "model": model,
                    "tool": mutation_tool,
                    "bias_type" : bias_type,
                    "mutant": len(df),
                    "fairness_violation": fairness_violation
                },
                ignore_index=True)

d
```
