# Create BTCs based on Country -> This code only work if the name for each country is one

```python
start = time.time()

identifier = "country"

mutant_text_1 = []
mutant_text_2 = []
prediction_1 = []
prediction_2 = []
identifier_1 = []
identifier_2 = []
template = []
original = []
for i in range(len(gb.size())) :
    data = gb.get_group(i)
    already_processed = []
    for index1, v1 in data.iterrows() :
        for index2, v2 in data.iterrows() :
            k1 = v1[identifier]
            k2 = v2[identifier]
            if k1 != k2 :
                k = k1 + "-" + k2
                if k1 > k2 :
                    k = k2 + "-" + k1
                if k not in already_processed :
                    already_processed.append(k)
                    p_1 = v1["prediction"]
                    p_2 = v2["prediction"]
                    if p_1 != p_2 :
                        mutant_text_1.append(v1["mutant"])
                        prediction_1.append(v1["prediction"])
                        identifier_1.append(k1)
                        mutant_text_2.append(v2["mutant"])
                        prediction_2.append(v2["prediction"])
                        identifier_2.append(k2)
                        template.append(v1["template"])
                        original.append(v1["original"])
                        

end = time.time()
print("Execution time: ", end-start)
```

```python
btc = pd.DataFrame(data={"mutant_1" : mutant_text_1, "mutant_2" : mutant_text_2, "prediction_1": prediction_1, "prediction_2" : prediction_2, "identifier_1": identifier_1, "identifier_2" : identifier_2, "template": template, "original": original})
btc
```

```python
btc = btc.sample(frac=1, random_state=123)

texts = []
templates = []
for index, rows in btc[:50].iterrows():
    texts.append(rows["original"])
    texts.append(rows["mutant_1"])
    texts.append(rows["mutant_2"])
    templates.append(rows["template"])
```
