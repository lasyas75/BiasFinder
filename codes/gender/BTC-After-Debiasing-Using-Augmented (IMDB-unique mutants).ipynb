# BTC

```python
import pandas as pd
import numpy as np
import math
import time
```

#### Biasfinder BTC

```python
df = pd.read_csv("../../data/btc/biasfinder/gender/original.csv")
df
```

```python
def read_txt(fpath):
    pred = []
    file = open(fpath)
    lines = file.readlines()
    for l in lines :
        pred.append(int(l))
    file.close()
    
    return pred
```

```python
output_dir = "btc/after_debiasing_using_augmented_unique_eec/biasfinder/"

result_dir = "../../result/" + output_dir + "/"

path = result_dir + "results_data.txt"

pred = read_txt(path)

print(len(pred))
```

```python
df["prediction_1"] = pred[:len(df)]
df["prediction_2"] = pred[len(df):]
```

```python
df
```

### Calculate BTC

```python
df["btc"] = df["prediction_1"] != df["prediction_2"]
df
```

```python
len(df[df["btc"] == True])
```
