```python
def markMutant(mutant, template) :

    chunks = []
    chunk = ""
    for word in template.split(" ") :
        if len(word) > 2 :
            if word[0] == "<" and word[-1] == ">" :
                if chunk != "" :
                    chunks.append(str(chunk[:-1]))
                else :
                    chunks.append("")
                chunk = ""

            else :
                chunk += word + " "
        else :
            chunk += word + " "
    if chunk != "" :
        chunks.append(str(chunk[:-1]))

#     print(chunks)
#     print()

    marked = str(mutant)

    word = chunks[0]
    end_index = marked.find(word) + len(word)
    if word == "" :
        marked = marked[:end_index] + '<' + marked[end_index:]
    else :
        marked = marked[:end_index+1] + '<' + marked[end_index+1:]
    # print("marked1: " + marked)
    # print()
    for i in range(1, len(chunks)):
        word = chunks[i]
        if i == len(chunks) - 1 :
            start_index = marked.find(word)
            if start_index != -1 :
                marked = marked[:start_index-1] + '>' + marked[start_index-1:]
    #             print("markedLast: " + marked)
    #             print()
        else :
            start_index = marked.find(word)
            if start_index != -1 :
                end_index = start_index + len(word)
                marked = marked[:end_index+1] + '<' + marked[end_index+1:]
                marked = marked[:start_index-1] + '>' + marked[start_index-1:]
    #             print("marked: " + marked)
    #             print()
    return marked
```

```python
mutant = "Charlie has shaped the face of British TV comedy. He has proved that you do not need wicked words or wit, a massive budget, a great deal of intelligence or even any intelligence to make something brilliant. Charlie is one of those characters who you just can't forget."
template = "<name> has shaped the face of British TV comedy. <pro-He> has proved that you do not need wicked words or wit, a massive budget, a great deal of intelligence or even any intelligence to make something brilliant. <name> is one of those characters who you just can't forget."
markMutant(mutant, template)
```
