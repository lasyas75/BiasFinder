```python
import pandas as pd
```

```python
imdb = pd.read_csv("../../data/imdb/train.csv", header=None, sep="\t", names=["label", "mutant"])
imdb
```

# Augment with IMDB

```python
import os

base_dir = "../../data/btc/biasfinder/gender/full/"

biasfinder = pd.read_csv(base_dir + "train.csv",sep="\t", names=["label", "mutant"])
biasfinder
```

```python
augmented_imdb_biasfinder = pd.concat([imdb, biasfinder])
augmented_imdb_biasfinder = augmented_imdb_biasfinder.sample(frac=1, random_state=123)
augmented_imdb_biasfinder
```

```python
augmented_imdb_biasfinder.to_csv("../../data/augmented/imdb_full_biasfinder/train.csv", index=None, header=None, sep="\t")
```

# Augment with EEC

```python
import os

base_dir = "../../data/btc/eec/gender/full/"

eec = pd.read_csv(base_dir + "train.csv", header=None, sep="\t", names=["label", "mutant", "text_id"])
eec = eec.drop(columns=["text_id"])
eec
```

```python
augmented_imdb_eec = pd.concat([imdb, eec])
augmented_imdb_eec = augmented_imdb_eec.sample(frac=1, random_state=123)
augmented_imdb_eec
```

```python
augmented_imdb_eec.to_csv("../../data/augmented/imdb_full_eec/train.csv", index=None, header=None, sep="\t")
```
