### Data Preparation for IMDB

`../data/imdb/train.csv` and `../data/imdb/test.csv` are already built for training. No need to change the data. This notebook is intended to produce a subset small data from the IMDB to make the code testing faster.

```python
import pandas as pd
import numpy as np
import math
import os
import random 
```

```python
df = pd.read_csv("../data/imdb/train.csv", header=None, sep="\t")
df
```

```python
dft = pd.read_csv("../data/imdb/test.csv", header=None, sep="\t")
dft
```

```python
# take the first 1000 rows
train = df.iloc[:1000]
test = dft.iloc[:1000]
```

```python
if not os.path.exists("../data/imdb_small/") :
    os.makedirs("../data/imdb_small/")

train.to_csv("../data/imdb_small/train.csv", index=None, header=None, sep="\t")
test.to_csv("../data/imdb_small/test.csv", index=None, header=None, sep="\t")
```
