#### Reading All Corpus from Database

```python
from os import listdir
from os.path import join, isfile

basedir = "../../data/gc_name/"
dirname = basedir + "raw/"

mpaths = []
fpaths = []

for f in listdir(dirname): 
    if isfile(join(dirname, f)) :
        if ".csv" == f[-4:] :
            if "Male" in f :
                mpaths.append(join(dirname, f))
            if "Female" in f :
                fpaths.append(join(dirname, f))
                
mpaths = sorted(mpaths)
fpaths = sorted(fpaths)
```

```python
from genderComputer import GenderComputer
gc = GenderComputer()

 

name_to_gender = {}
for country in gc.nameLists.keys():
    for gender, names in gc.nameLists[country].items():
        for name in names:
            if name in name_to_gender and  name_to_gender[name] != gender:
                name_to_gender[name] = 'invalid'
            else:
                name_to_gender[name] = gender

 


result = [(name, gender) for name, gender in name_to_gender.items() if gender != 'invalid']
```

```python
# selected country if the name for each male and female are more than N
n = 5

def getTheNumberOfNames(path) :
    d = read_data(path)
    return len(d)

selected_countries = []

for m, f in zip(mpaths, fpaths) :
    mcountry = m[len(dirname):-12]
    fcountry = f[len(dirname):-14]
    if mcountry == fcountry :
        if getTheNumberOfNames(m) >= 2 * n and getTheNumberOfNames(f) >= 2 * n : #add buffer
            selected_countries.append(mcountry)

selected_countries
```

### Get Unique Male Female Name

```python
m_unique_names = set()
for path in mpaths :
    d = read_data(path)
    for index, row in d.iterrows() :
        name = str(rows["Name"]).title()
        m_unique_names.add(name)    
```

```python
f_unique_names = set()
for path in fpaths :
    d = read_data(path)
    for index, row in d.iterrows() :
        name = str(rows["Name"]).title()
        f_unique_names.add(name)
```

### Get n Names for each country

```python
# selected country if the name for each male and female are more than N
n = 5
seed = 123456

df = pd.DataFrame(columns=["name", "gender", "country"])

for mpath, fpath in zip(mpaths, fpaths) :
    mcountry = mpath[len(dirname):-12]
    fcountry = fpath[len(dirname):-14]

    if mcountry == fcountry :
        country = mcountry

        mdf = read_data(mpath)
        mnames = []
        i = 0
        for index, rows in mdf.sample(frac=1, random_state=seed).iterrows() :
            name = str(rows["Name"]).title()
            if name not in f_unique_names :
                mnames.append(name)
                i += 1
                if i == n :
                    break
        
        
        fdf = read_data(fpath)
        fnames = []
        i = 0
        for index, rows in fdf.sample(frac=1, random_state=seed).iterrows() :
            name = str(rows["Name"]).title()
            if name not in m_unique_names :
                fnames.append(name)
                i += 1
                if i == n :
                    break
        
        if len(mnames) >= n and len(fnames) >= n :
            gender = "male"
            for name in mnames :
                df = df.append({"name": name, "gender": gender, "country" : country}, ignore_index=True)

            gender = "female"
            for name in fnames :
                df = df.append({"name": name, "gender": gender, "country" : country}, ignore_index=True)
```
